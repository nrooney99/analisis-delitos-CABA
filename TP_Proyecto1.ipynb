```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

comunasUrl="https://cdn.buenosaires.gob.ar/datosabiertos/datasets/comunas/CABA_comunas.geojson"
barriosUrl="https://cdn.buenosaires.gob.ar/datosabiertos/datasets/barrios/barrios.geojson"

map_dataC = gpd.read_file(comunasUrl)
delitos= pd.Series([9560,4438,7653,7900,5083,4094,6846,5801,5353,4285,3965,4772,5706,7845,5167],index=[14,0,11,8,9,1,10,13,12,2,3,4,7,5,6])
map_dataC=map_dataC.assign(DELITOS=delitos)

map_dataB = gpd.read_file(barriosUrl)
```

```python
# Control del tamaño de la figura del mapa
fig, ax = plt.subplots(figsize=(10, 10))
 
# Control del título y los ejes
ax.set_title('Delitos por comuna en CABA, 2021', 
             pad = 20, 
             fontdict={'fontsize':20, 'color': '#4873ab'})
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')

# Añadir la leyenda separada del mapa
from mpl_toolkits.axes_grid1 import make_axes_locatable
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.2)
 
# Mostrar el mapa finalizado
map_dataC.plot(column='DELITOS', cmap='Reds', ax=ax,legend=True, cax=cax, zorder=5)
```
